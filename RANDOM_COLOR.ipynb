#                    **Random Colour Generator**

```python
import tkinter as tk
import random

root = tk.Tk()
root.geometry("300x200")

colors=["red","blue","green","yellow","pink","orange"]

def change_color():
    root.config(bg=random.choice(colors))

button=tk.Button(root,text="Change Color",command=change_color)
button.pack(pady=80)

root.mainloop()
```
